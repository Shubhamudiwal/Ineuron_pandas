```python
#1 . Find out how many male and female passenger was onboarded .
print('No.of male',max(df[df['Sex']=='male'].count()))
print('No.of female',max(df[df['Sex']=='female'].count()))
```

```python
#2. how many survived we have .
print('No of survived',max(df[df['Survived']==1].count()))
```

```python
#3. how many casuality we have 
print('No of casuality',max(df[df['Survived']==0].count()))
```

```python
#4 . what is name of a person who is the eldest one . 
df[df['Age']==max(df['Age'])]
```

```python
#6. how many person we have whose name starts with "s"
df[df['Name'].str.startswith('S')]
```

```python
#7. try to create a new column which is a summation  of "SibSp" and "parch"
df[df['sumation']==df['SibSp']+df['Parch']]
```

```python
#how many person do we have below age of 25 .
df[df['Age']<25]
```

```python
#9 . how many person died whose age was less then 40 
df.loc[(df['Age']==40) & (df['Survived']==1)]
```

```python
#from a  cabin column seperate text and numeric value . 
for i in df.loc[df['Cabin'].notnull()]['Cabin']:
    print(i.extract('([A-Z]+)'))
```
